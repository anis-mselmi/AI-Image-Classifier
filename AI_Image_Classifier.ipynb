Auto-generate three simple sample images (one each for apple, banana, and orange) using PIL. Save these images into their respective 'dataset' subfolders.

## Task Completed

All steps of the project have been successfully completed:

```python
import os
from PIL import Image, ImageDraw

project_dir = 'project'
class_names = ['apple', 'banana', 'orange']

img_size = (128, 128)

print("Generating sample images...")

for class_name in class_names:
    img = Image.new('RGB', img_size, color = 'white')
    d = ImageDraw.Draw(img)

    if class_name == 'apple':
        d.ellipse((30, 30, 98, 98), fill='red', outline='black')
        d.line((64, 20, 64, 30), fill='brown', width=3)
        d.ellipse((70, 25, 80, 35), fill='green')
    elif class_name == 'banana':
        d.arc((20, 40, 100, 120), start=0, end=180, fill='yellow', width=20)
        d.ellipse((10, 50, 40, 70), fill='yellow')
        d.ellipse((80, 50, 110, 70), fill='yellow')
    elif class_name == 'orange':
        d.ellipse((30, 30, 98, 98), fill='orange', outline='black')

    save_dir = os.path.join(project_dir, 'dataset', class_name)
    image_filename = f"{class_name}_01.png"
    image_path = os.path.join(save_dir, image_filename)

    os.makedirs(save_dir, exist_ok=True)

    img.save(image_path)
    print(f"Generated and saved: {image_path}")

print("All sample images generated successfully.")
```

## Load Dataset

### Subtask:
Use TensorFlow's `ImageDataGenerator` to load the generated sample images from the 'dataset' directory, preparing them for training and validation.


**Reasoning**:
To load the generated sample images for training, `ImageDataGenerator` is the appropriate tool from TensorFlow/Keras. It allows for efficient loading of images from directories, applies preprocessing steps like rescaling, and handles batching. The `flow_from_directory` method automatically infers class labels from subdirectories, which is convenient for our project structure. Rescaling pixel values to `[0, 1]` is a standard preprocessing step for neural networks.

**Reasoning**:
I will import `ImageDataGenerator`, define image parameters, instantiate the generator with rescaling, and use `flow_from_directory` to load images from the dataset. Finally, I'll print the inferred class names for verification.



```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define image dimensions and batch size
img_height, img_width = 128, 128
batch_size = 32

# Path to the dataset directory
dataset_dir = 'project/dataset'

# Instantiate ImageDataGenerator with rescaling
train_datagen = ImageDataGenerator(rescale=1./255)

# Use flow_from_directory to load images
train_generator = train_datagen.flow_from_directory(
    dataset_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

# Print the class names inferred by the generator
print("Class labels:", train_generator.class_indices)
print(f"Found {train_generator.samples} images belonging to {train_generator.num_classes} classes.")
```

## Build and Compile CNN Model

### Subtask:
Construct a small Convolutional Neural Network (CNN) using TensorFlow/Keras suitable for basic image classification, then compile it with an appropriate optimizer, loss function, and metrics.


**Reasoning**:
I need to construct the CNN model as specified in the instructions, then compile it and print its summary. This involves importing necessary Keras layers, defining the sequential model architecture, and finally compiling it with the given optimizer, loss, and metrics.



```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(train_generator.num_classes, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("Model Summary:")
model.summary()
```

**Reasoning**:
The previous code block successfully built and compiled the CNN model. The warning about `input_shape` in `Sequential` models is non-critical and the model summary was printed as expected. The next logical step according to the task is to train the model.



```python
print("Training the model...")

# Train the model for 5 epochs
history = model.fit(
    train_generator,
    epochs=5,
    verbose=1
)

print("Model training complete.")
```

**Reasoning**:
The previous step successfully trained the model. The next logical step is to save the trained model to the specified path `models/model.h5` for future use, as outlined in the main task description.



```python
import os

# Define the path to save the model
model_save_path = os.path.join('project', 'models', 'model.h5')

# Save the trained model
model.save(model_save_path)

print(f"Model saved successfully to {model_save_path}")
```

**Reasoning**:
The model has been trained and saved. The next logical step, as per the main task description, is to generate and save plots of the training accuracy and loss to the `results/` directory using the `history` object obtained during training.



```python
import matplotlib.pyplot as plt
import os

# Define the results directory
results_dir = os.path.join('project', 'results')

# Plot training accuracy
plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'], label='Accuracy')
plt.title('Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
accuracy_plot_path = os.path.join(results_dir, 'training_accuracy.png')
plt.savefig(accuracy_plot_path)
print(f"Training accuracy plot saved to {accuracy_plot_path}")
plt.show()

# Plot training loss
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Loss')
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
loss_plot_path = os.path.join(results_dir, 'training_loss.png')
plt.savefig(loss_plot_path)
print(f"Training loss plot saved to {loss_plot_path}")
plt.show()

print("Plots generated and saved successfully.")
```

**Reasoning**:
The previous steps covered model training, saving, and plotting. The next logical step, as per the main task description, is to create a prediction script that loads the saved model, prepares a test image for prediction, and displays the predicted class.



```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import os

model_path = os.path.join('project', 'models', 'model.h5')

loaded_model = tf.keras.models.load_model(model_path)
print(f"Model loaded successfully from {model_path}")

if 'train_generator' in locals():
    class_indices = train_generator.class_indices
else:
    class_indices = {'apple': 0, 'banana': 1, 'orange': 2}

idx_to_class = {v: k for k, v in class_indices.items()}

print("Available classes for prediction:", list(class_indices.keys()))

def predict_image(img_path):
    img = image.load_img(img_path, target_size=(img_height, img_width))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0

    predictions = loaded_model.predict(img_array)
    predicted_class_idx = np.argmax(predictions[0])
    predicted_class_name = idx_to_class[predicted_class_idx]
    confidence = np.max(predictions[0])

    print(f"\nPredicting for: {img_path}")
    print(f"Predicted class: {predicted_class_name} (Confidence: {confidence:.2f})")
    print(f"Raw predictions: {predictions[0]}")
    return predicted_class_name, confidence

test_image_path = os.path.join('project', 'dataset', 'apple', 'apple_01.png')

if os.path.exists(test_image_path):
    predict_image(test_image_path)
else:
    print(f"Error: Test image not found at {test_image_path}")
    print("Please ensure the image path is correct or create a new image for testing.")

print("Prediction script executed.")
```
